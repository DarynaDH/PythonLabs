```python
import time

def delay_print(s):
    for c in s:
        print(c, end='')
        time.sleep(0.0)
    print()

answers_array = []


quest_index = 0

delay_print("Вітаю на першому рівні! До тебе підходить невідоме створіння, схоже на гнома, і каже:\n'Привіт, незнайомцю. Якщо ти хочеш пройти, то тобі потрібно відгадати загадку. Якщо ти не хочеш, то я тебе вб'ю.'")

print("Варіанти відповідей:\n1) 'Так, я хочу відгадати загадку.'\n2) 'Ні, я не хочу відгадувати загадку. (почати драку)'")

answer = input("Введіть відповідь: ")
answers_array.append(answer)

match answer:
    case "1":
        delay_print("Гном: 'Добре, тоді почнемо. Якщо ти відгадаєш, то я тебе відпущу.'")
    case "2":
        delay_print("Гном: 'Ти вирішив битися? Тоді почнемо.'")
    case _:
        delay_print("Гном: 'Я не розумію тебе. Якщо ти не зможеш відповісти, то я тебе вб'ю.'")


quest_index = 1
previous_answer = answers_array[quest_index - 1]

match previous_answer:
    case "1":
        delay_print("Гном: 'У мами Івана було три сини. Першого звали Петро, другого – Василь. Як звали третього сина?'")
        print("Варіанти відповідей:\n1) 'Іван'\n2) 'Петро'\n3) 'Василь'")
        answer = input("Введіть відповідь: ")
    case "2":
        print("*Гном розмахується кулаком і ударяє вас по коліну.*")
        delay_print("Гном: 'Н-Н-Н-А-А-А!!!'")
        print("*Ви відчуваєте сильний біль в коліні, але він не надто сильний*")
        print("1. *пнути гнома з усієї сили ногою*\n2. Будь ласка!!! Не бий! Я відповім на твою загадку!!!")
        answer = input("Введіть відповідь: ")
    case _:
        delay_print("Гном: 'Шо ти там бормочеш??!!'")


answers_array.append(answer)


quest_index = 2
previous_answer = answers_array[quest_index - 1]

match previous_answer:
    case "1":
        delay_print("Гном: 'Вірно! Ти розумний!'")
    case "2":
        delay_print("Гном: 'Ти маєш сміливість. Але ти вже не зможеш продовжити свій шлях.'")
    case _:
        delay_print("Гном: 'Ти щось плутаєш. Якщо не зможеш відповісти на наступне питання, то ти виграєш.'")


answers_array.append(answer)

quest_index = 3
previous_answer = answers_array[quest_index - 1]

match previous_answer:
    case "1":
        delay_print("Гном: 'На другому рівні перед тобою вибір - дві дороги. Одна веде в ліс, а інша веде до річки. Куди підеш?'")
        print("Варіанти відповідей:\n1) 'В ліс'\n2) 'До річки'")
        answer = input("Введіть відповідь: ")
    case "2":
        delay_print("Гном: 'Дорогою ти потрапив у яму та злавам ногу що заповільнило тебе'")
    case _:
        delay_print("Гном: 'Не гай часу, обирай! Відповідай!'")


answers_array.append(answer)

quest_index = 4
previous_answer = answers_array[quest_index - 1]

match previous_answer:
    case "1":
        delay_print("Гном: 'Ти обрав в ліс. Перед тобою стоїть річка і ти бачиш місток. Перейдеш річку чи повернешся?'")
        print("Варіанти відповідей:\n1) 'Перейду річку'\n2) 'Повернуся'")
        answer = input("Введіть відповідь: ")
    case "2":
        delay_print("Гном: 'Ти  занадто довго вагався, і зараз вже пізно.'")
    case _:
        delay_print("Гном: 'Не гай часу, обирай! Відповідай!'")


answers_array.append(answer)


quest_index = 5
previous_answer = answers_array[quest_index - 1]

match previous_answer:
    case "1":
        delay_print("Гном: 'Ти обрав перейти річку. Річка мілка, ти успішно перейшов її, вітаю!'")
    case "2":
        delay_print("Гном: 'Ти повернувся але настала ніч, ти заблукав у лісі'")
    case _:
        delay_print("Гном: 'Не гай часу, обирай! Відповідай!'")


answers_array.append(answer)

answer_sum = 0
for e in answers_array:
    answer_sum += int(e)

if answer_sum == 6:
    delay_print("Гном: 'Ти успішно пройшов гру та досяг останнього рівня. Ти переміг! Вітаю!'")
else:
    delay_print("Гном: 'Ти не сподобався гному й він тебе з'їв. Гра завершена. Гном переміг.'")
```
