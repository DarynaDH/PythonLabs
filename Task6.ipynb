```python
from datetime import datetime

log_history = []


def logger(func):
    def wrapper(*args, **kwargs):
        try:
            start_time = datetime.now()
            result = func(*args, **kwargs)
            log_entry = {
                'log type': 'Good',
                'time': start_time,
                'function_name': func.__name__,
                'arguments': [args, kwargs],
                'result': result,
            }
        except Exception as e:
            log_entry = {
                'log type': 'Error',
                'time': datetime.now(),
                'function_name': func.__name__,
                'arguments': [args, kwargs],
                'error_message': str(e),
            }
        with open('function_logs.txt', 'a') as f:
            f.write(str(log_entry) + '\n')

    return wrapper


def get_logs(log_type=None):
    with open('function_logs.txt', 'r') as log_file:
        data = log_file.read().split('\n')
    for log in data:
        if log_type is None:
            yield log
        else:
            if log_type in log:
                yield log


@logger
def divide(a, b):
    if b == 0:
        raise Exception("Division by zero")
    return a / b


@logger
def multiply(a, b):
    return a * b


@logger
def addition(a, b):
    return a + b


while True:
    print("Меню:")
    print("1. Виконати функцію ділення")
    print("2. Виконати функцію множення")
    print("3. Виконати функцію додавання")
    print("4. Вивести всі логи")
    print("5. Вивести логи типу 'Good'")
    print("6. Вивести логи типу 'Error'")
    print("7. Вийти")

    choice = input("Виберіть опцію: ")

    if choice == '1':
        a = float(input("Введіть перше число (a): "))
        b = float(input("Введіть друге число (b): "))
        divide(a, b)
    elif choice == '2':
        a = float(input("Введіть перше число (a): "))
        b = float(input("Введіть друге число (b): "))
        multiply(a, b)
    elif choice == '3':
        a = float(input("Введіть перше число (a): "))
        b = float(input("Введіть друге число (b): "))
        addition(a, b)
    elif choice == '4':
        logs = get_logs()
        for entry in logs:
            print(entry)
    elif choice == '5':
        log_type = 'Good'
        logs = get_logs(log_type)
        for entry in logs:
            print(entry)
    elif choice == '6':
        log_type = 'Error'
        logs = get_logs(log_type)
        for entry in logs:
            print(entry)
    elif choice == '7':
        break
    else:
        print("Невірний вибір. Спробуйте ще раз.")

```
