```python
class SushiRestaurant:
    def __init__(self):
        self.menu = {
            "Ролл Каліфорнія": 200,
            "Суші з лососем": 250,
            "Суші з креветками": 275,
        }

    def prepare_order(self, item):
        if item in self.menu:
            return f"Приготовано {item}"
        else:
            return f"Даний {item} відсутній у меню суші-ресторану."


class Pizzeria:
    def __init__(self):
        self.menu = {
            "Піца Маргарита": 210,
            "Піца Пепероні": 255,
            "Піца Гавайська": 270,
        }

    def prepare_order(self, item):
        if item in self.menu:
            return f"Приготовано {item}"
        else:
            return f"Даний {item} відсутній у меню піцерії."


class Steakhouse:
    def __init__(self):
        self.menu = {
            "Стейк з яловичини": 355,
            "Стейк з курки": 280,
            "Стейк з свинини": 320,
            "Стейк з індички": 290,
        }

    def prepare_order(self, item):
        if item in self.menu:
            return f"Приготовано {item}"
        else:
            return f"Даний {item} відсутній у меню стейк хаусу."


class Order:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def place_order(self, restaurant):
        for item in self.items:
            print(restaurant.prepare_order(item))



orders_list = []

def main():
    sushi_restaurant = SushiRestaurant()
    pizzeria = Pizzeria()
    steakhouse = Steakhouse()

    print("Виберіть ресторан:\n")
    print("1. Суші-ресторан")
    print("2. Піцерія")
    print("3. Стейк-хауз")

    choice = input("Введіть номер ресторану: \n")

    if choice == "1":
        restaurant = sushi_restaurant
    elif choice == "2":
        restaurant = pizzeria
    elif choice == "3":
        restaurant = steakhouse
    else:
        print("Неправильний вибір ресторану. Спробуйте ще")
        return main()

    order = Order()
    if choice == "1":
        for key, value in sushi_restaurant.menu.items():
            print(key, '-', value)
    elif choice == "2":
        for key, value in pizzeria.menu.items():
            print(key, '-', value)
    elif choice == "3":
        for key, value in steakhouse.menu.items():
            print(key, '-', value)

    while True:
        item = input("Введіть страву (або 'закінчити' для завершення замовлення): ")

        if item.lower() == "закінчити":
            break

        order.add_item(item)

    print("Замовлення:\n")
    order.place_order(restaurant)
    orders_list.extend(order.items)

    answ = str(input("\nЧи бажаєте зробити ще одне замовлення?"))

    if answ.lower() == "так":
        return main()
    if answ.lower() == "ні":
        print('\nДякуємо за ваше замовлення!\n')
        for e in orders_list:
            print(e)
        return


if __name__ == "__main__":
    main()
```
